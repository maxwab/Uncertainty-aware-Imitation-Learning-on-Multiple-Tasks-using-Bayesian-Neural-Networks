```python
import _pickle as pickle
import numpy as np
from Settings import *

from plots import plotWithMasses
```

```python
file_name = './logs/SlidingBlock/False_2.pkl'
with open(file_name, 'rb') as f:
    data = pickle.load(f)
    
average_episodic_cost_over_all_masses = []
average_episodic_deviation_over_all_masses = []
average_episodic_action_over_all_masses = []
average_episodic_maximum_action_over_all_masses = []
average_episodic_minimum_action_over_all_masses = []
all_position_gains = []
all_velocity_gains = []

for each_mass in ALL_BLOCK_MASSES_TO_VALIDATE:
    each_mass = str(each_mass)
    episodic_cost_over_this_mass = 0.
    episodic_deviation_over_this_mass = 0.
    episodic_action_over_this_mass = 0.
    episodic_maximum_over_this_mass = 0.
    episodic_minimum_over_this_mass = 0.
    for each_initial_state in INITIALIZATION_STATES:
        each_initial_state = str(each_initial_state)
        episodic_cost_over_this_mass = episodic_cost_over_this_mass + np.sum(data[each_mass][each_initial_state][COST_LOG_KEY])
        episodic_deviation_over_this_mass = episodic_deviation_over_this_mass + np.sum(data[each_mass][each_initial_state][DEVIATION_LOG_KEY])
        episodic_action_over_this_mass = episodic_action_over_this_mass + np.sum(data[each_mass][each_initial_state][ACTION_TAKEN_LOG_KEY])
        episodic_maximum_over_this_mass = episodic_maximum_over_this_mass + np.sum(data[each_mass][each_initial_state][MAXIMUM_ACTION_LOG_KEY])
        episodic_minimum_over_this_mass = episodic_minimum_over_this_mass + np.sum(data[each_mass][each_initial_state][MINIMUM_ACTION_LOG_KEY])
    average_episodic_cost_over_all_masses.append(episodic_cost_over_this_mass/INITIALIZATION_STATES.shape[0])
    average_episodic_deviation_over_all_masses.append(episodic_deviation_over_this_mass/INITIALIZATION_STATES.shape[0])
    average_episodic_action_over_all_masses.append(episodic_action_over_this_mass/INITIALIZATION_STATES.shape[0])
    average_episodic_maximum_action_over_all_masses.append(episodic_maximum_over_this_mass/INITIALIZATION_STATES.shape[0])
    average_episodic_minimum_action_over_all_masses.append(episodic_minimum_over_this_mass/INITIALIZATION_STATES.shape[0])
    all_position_gains = data[each_mass][POSITION_GAIN_KEY]
    all_velocity_gains = data[each_mass][VELOCITY_GAIN_KEY]
    
plotWithMasses(average_episodic_cost_over_all_masses, 'Rewards', BLOCK_MASSES_TO_TRAIN_ON_3)
plotWithMasses(average_episodic_deviation_over_all_masses, 'Uncertainty', BLOCK_MASSES_TO_TRAIN_ON_3)
plotWithMasses(all_position_gains, 'Position Gain', BLOCK_MASSES_TO_TRAIN_ON_3)
plotWithMasses(all_velocity_gains, 'Velocity Gain', BLOCK_MASSES_TO_TRAIN_ON_3)
```
